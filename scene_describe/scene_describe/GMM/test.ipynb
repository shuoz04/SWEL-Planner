```python
import numpy as np
import matplotlib.pyplot as plt


def plot(y1,y2,y3,y_axix):
    percentages = np.arange(20, 100, 10)   # [20, 30, ..., 90]
    x = np.insert(percentages, 0, 0)         # [0, 20, 30, ..., 90]

    # 示例三组9维 y 值数组（请替换成你的数据）

    # 绘图
    plt.figure(figsize=(8, 5))
    plt.plot(x, y1, marker='o', linestyle='-', label='pos_only')
    plt.plot(x, y2, marker='s', linestyle='--', label='orien_only')
    plt.plot(x, y3, marker='^', linestyle='-.', label='both')

    plt.xlabel('Similarity')
    plt.ylabel(y_axix)
    plt.title('Line Plot of Three Groups of Y Values')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
plot(valid_point_pos_only,valid_point_orien_only,valid_point_both,'Valid Points')
```

```python
plot(edge_in_total_pos_only,edge_in_total_orien_only,edge_in_total_both,'Edge in Total')
```
